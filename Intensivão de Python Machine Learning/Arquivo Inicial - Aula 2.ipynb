# Análise de Dados com Python

### Desafio:

Você trabalha em uma empresa de telecom e tem clientes de vários serviços diferentes, entre os principais: internet e telefone.

O problema é que, analisando o histórico dos clientes dos últimos anos, você percebeu que a empresa está com Churn de mais de 26% dos clientes.

Isso representa uma perda de milhões para a empresa.

O que a empresa precisa fazer para resolver isso?

```python
import pandas as pd

tabela = pd.read_csv(r"C:\Users\arthu\Downloads\telecom_users.csv")
tabela
```

```python
tabela.columns
```

```python
tabela.info()
```

```python
# Identificar tabelas que não te ajudaram pra nd
# Colunas "Unnamed: 0, Codigo" Não Servem Pra Nada
tabela = tabela.drop("Unnamed: 0", axis=1) # Não mostro info relevante
tabela = tabela.drop("Codigo", axis=1) # Tava tudo vazia
tabela
```

```python
tabela.info()
```

```python
# Tratar os Dados
tabela["TotalGasto"] = pd.to_numeric(tabela["TotalGasto"], errors="coerce")# vai deixar os erros com valor vazio (NaN)
tabela.info()
```

```python
# Tratar Infomações Vazias
tabela = tabela.dropna(how="all", axis=1) # Excluir colunas completamente vazias #Aki a coluna "Codigo" iria de vala, mais ja foi tratada la em cima
tabela = tabela.dropna(how="any", axis=0) # Excluir linhas vazias
tabela.info()
```

```python
# Informações Tratadas ==  Começar Analise de Dados
# a empresa está com Churn de mais de 26% dos clientes. Isso representa uma perda de milhões para a empresa
# Coluna "Churn" é a coluna dos cancelamentos

print(tabela["Churn"].value_counts())             # Isso aki é só pra formatar bunitin
print(tabela["Churn"].value_counts(normalize=True).map("{:.1%}".format)) # Isso vai pegar a contagem de dados e transformala em percentual
```

```python
# Criar graficos para facilitar a aniise
import plotly.express as px

                    #( tabela, coluna, comparação de outracoluna em relação a cor)
grafico = px.histogram(tabela, x="MesesComoCliente", color="Churn", text_auto=True)
# grafico.show() # os graficos tão bugando no jupyter, ai só deixa comentei e dps discomenta pra ver
```
